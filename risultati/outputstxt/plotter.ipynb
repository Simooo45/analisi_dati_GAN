```python
import matplotlib.pyplot as plt
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Funzione per caricare i file di testo
def load_data(file_path):
    return pd.read_csv(file_path, delim_whitespace=True, header=None, names=['y'])

# Lista dei file di testo
data_files = [r'D:\D\gitRepos\analisi_dati_GAN\risultati\outputstxt\outputstxt\outputs\run3L_200_1024_16\caloChallenge\Baseline_pions_lwtnn\All\checkpoints\pions\checkpoints_eta_20_25\chi2.txt', r'D:\D\gitRepos\analisi_dati_GAN\risultati\outputstxt\outputstxt\outputs\run3L_200_1024_16\caloChallenge\Baseline_pions_lwtnn\All\checkpoints\pions\checkpoints_eta_20_25\chi2.txt', r'D:\D\gitRepos\analisi_dati_GAN\risultati\outputstxt\outputstxt\outputs\run3L_200_1024_16\caloChallenge\Baseline_pions_lwtnn\All\checkpoints\pions\checkpoints_eta_20_25\chi2.txt']  # Modifica con i tuoi file

# Carica i dati in un dizionario
data = {file: load_data(file) for file in data_files}

# Crea un plot
fig, ax = plt.subplots()

# Crea i plot per ciascun file
plots = {}
for file, df in data.items():
    plots[file], = ax.plot(df.index, df['y'], label=file, alpha=0.7, marker='o', linestyle='')

# Funzione per aggiornare il grafico
def update_plot(change):
    file = change.owner.description
    plots[file].set_visible(change.new)
    plt.draw()

# Crea le checkbox
checkboxes = [widgets.Checkbox(description=file, value=True, layout=widgets.Layout(width='100px')) for file in data_files]
for checkbox in checkboxes:
    checkbox.observe(update_plot, 'value')
    display(checkbox)

ax.legend()
plt.show()
```
